Burada 3 farklı algoritma ile fibonacci serisi işlenecek ve bir "Timer Decorator" kodlanıp algoritmalar arası performans farkı test edilecektir.

-----------------------------------------------------------------------------------------------------

TIMER DECORATOR

```python
def timer(fn):
    from time import perf_counter
    from functools import wraps
    
    @wraps(fn)
    def inner(*args,**kwargs):
        start = perf_counter()
        result = fn(*args,**kwargs)
        finish = perf_counter()
        elapsed = finish - start
        
        args_ = [str(a) for a in args]
        kwargs_ = ["{0} = {1}".format(k,v) for (k,v) in kwargs.items()]
        all_args = args_+kwargs_
        args_str = ','.join(all_args)
        
        print("{0}({1}) took {2:.6f}s to run".format(fn.__name__, args_str, elapsed))
        
        return result
    return inner
```

1.ALGORİTMA -> RECURSION

```python
def fib_rec_calc(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fib_rec_calc(n-1) + fib_rec_calc(n-2)   
    
# Eğer "timer decorator" bu fonksiyona yerleştirilseydi özyinelemenin her adımı ekrana yansıtılacaktı. Bunun önüne geçmek için bir tampon 
# görevi görecek fonksiyon yaratılmalı. (fib_rec)
```

```python
@timer
def fib_rec(n):
    return fib_rec_calc(n)
```

```python
fib_rec(30)
```

2.ALGORİTMA -> LOOP

```python
@timer
def fib_loop(n):
    a, b = 1, 1
    for _ in range(3, n+1):
        a, b = a+b, a
    return a        
```

```python
fib_loop(30)
```

3.ALGORİTMA -> REDUCE FUNCTION

```python
from functools import reduce

@timer
def fib_red(n):
    start = (1,1)
    itr = range(n-2)
    fib_n = reduce(lambda init, dummy:(init[1], init[0] + init[1]), itr, start)
    return fib_n[1]
```

```python
fib_red(30)
```
